```python
#MasVnrArea
data['MasVnrArea'] = data['MasVnrArea'].fillna(0)

#GarageYrBlt
data['GarageYrBlt'] = data['GarageYrBlt'].fillna(int(data['GarageYrBlt'].mean()))

#LotShape
data['LotShape'] = data['LotShape'].map({'IR3': 1, 'IR2': 2, 'IR1': 3, 'Reg': 4})

#Neighborhood
one_hot_encoding = pd.get_dummies(data['Neighborhood'])
one_hot_encoding = one_hot_encoding.rename( columns = { x:'Neighborhood-'+ x for x in one_hot_encoding.columns })
data = data.drop('Neighborhood', axis=1)
data = pd.concat( [one_hot_encoding, data], axis = 1 )

#BldgType
one_hot_encoding = pd.get_dummies(data['BldgType'])
one_hot_encoding = one_hot_encoding.rename( columns = { x:'BldgType-'+ x for x in one_hot_encoding.columns })
data = data.drop('BldgType', axis=1)
data = pd.concat( [one_hot_encoding, data], axis = 1 )

#MasVnrType
one_hot_encoding = pd.get_dummies(data['MasVnrType'])
one_hot_encoding = one_hot_encoding.rename( columns = { x:'MasVnrType-'+ x for x in one_hot_encoding.columns })
data = data.drop('MasVnrType', axis=1)
data = pd.concat( [one_hot_encoding, data], axis = 1 )

#ExterQual
data['ExterQual'] = data['ExterQual'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
data['ExterQual'] = data['ExterQual'].fillna(int(data['ExterQual'].mean()))

#BsmtQual
data['BsmtQual'] = data['BsmtQual'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
data['BsmtQual'] = data['BsmtQual'].fillna(int(data['BsmtQual'].mean()))

#BsmtExposure
data['BsmtExposure'] = data['BsmtExposure'].map({'Gd': 4, 'Av': 3, 'Mn': 2, 'No': 1})
data['BsmtExposure'] = data['BsmtExposure'].fillna(int(data['BsmtExposure'].mean()))

#HeatingQC
data['HeatingQC'] = data['HeatingQC'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
data['HeatingQC'] = data['HeatingQC'].fillna(int(data['HeatingQC'].mean()))

#KitchenQual
data['KitchenQual'] = data['KitchenQual'].map({'Ex': 5, 'Gd': 4, 'TA': 3, 'Fa': 2, 'Po': 1})
data['KitchenQual'] = data['KitchenQual'].fillna(int(data['KitchenQual'].mean()))

#GarageFinish
data['GarageFinish'] = data['GarageFinish'].map({'Fin': 1, 'RFn': 0.8, 'Unf': 0.3})
data['GarageFinish'] = data['GarageFinish'].fillna(int(data['GarageFinish'].mean()))

#drop掉的放一起
data = data.drop(['Id'],axis=1)         
data = data.drop(['MSZoning'],axis=1)    
data = data.drop(['LotFrontage'],axis=1)
data = data.drop(['Street'],axis=1)    
data = data.drop(['Alley'],axis=1)
data = data.drop(['LandContour'],axis=1)
data = data.drop(['Utilities'], axis=1)
data = data.drop(['LotConfig'],axis=1)
data = data.drop(['LandSlope'],axis=1)
data = data.drop(['Condition1'],axis=1)
data = data.drop(['Condition2'],axis=1)
data = data.drop(['HouseStyle'], axis=1)
data = data.drop(['RoofStyle'],axis=1)
data = data.drop(['RoofMatl'],axis=1)
data = data.drop(['Exterior1st'],axis=1)
data = data.drop(['Exterior2nd'],axis=1)
data = data.drop(['ExterCond'],axis=1)
data = data.drop(['Foundation'], axis=1)
data = data.drop(['BsmtCond'],axis=1)
data = data.drop(['BsmtFinType1'],axis=1)
data = data.drop(['BsmtFinType2'],axis=1)
data = data.drop(['Heating'],axis=1) 
data = data.drop(['CentralAir'],axis=1)
data = data.drop(['Electrical'],axis=1)
data = data.drop(['Functional'],axis=1)
data = data.drop('FireplaceQu', axis=1)
data = data.drop(['GarageType'], axis=1)
data = data.drop(['GarageQual'],axis=1)
data = data.drop(['GarageCond'],axis=1)
data = data.drop(['PavedDrive'],axis=1)
data = data.drop(['PoolQC'], axis=1)
data = data.drop(['Fence'],axis=1)
data = data.drop(['MiscFeature'], axis=1)
data = data.drop(['SaleType'],axis=1)
data = data.drop(['SaleCondition'], axis=1)
data = data.drop(['MiscVal'], axis=1)
data = data.drop(['PoolArea'], axis=1)
data = data.drop(['ScreenPorch'], axis=1)
data = data.drop(['3SsnPorch'], axis=1)
data = data.drop(['EnclosedPorch'], axis=1)
data = data.drop(['KitchenAbvGr'], axis=1)
data = data.drop(['BsmtHalfBath'], axis=1)
data = data.drop(['LowQualFinSF'], axis=1)
data = data.drop(['BsmtFinSF2'], axis=1)
data = data.drop(['ExterQual'], axis=1)
data = data.drop(['BsmtQual'], axis=1)
data = data.drop(['BsmtExposure'], axis=1)
data = data.drop(['HeatingQC'], axis=1)
data = data.drop(['2ndFlrSF'], axis=1)
data = data.drop(['BsmtFullBath'], axis=1)
data = data.drop(['FullBath'], axis=1)
data = data.drop(['HalfBath'], axis=1)
data = data.drop(['BedroomAbvGr'], axis=1)
data = data.drop(['KitchenQual'], axis=1)
data = data.drop(['Fireplaces'], axis=1)
data = data.drop(['GarageFinish'], axis=1)
data = data.drop(['GarageCars'], axis=1)
data = data.drop(['WoodDeckSF'], axis=1)
data = data.drop(['OpenPorchSF'], axis=1)
data = data.drop(['MSSubClass'], axis=1)
data = data.drop(['BsmtFinSF1'], axis=1)
```

#### [ 4. 將資料切割成訓練集 70%,預測集 30% ]
##### 在下面這個cell 先把資料切好
##### 把『 normalize後的資料 』也在這邊先切好

```python
from sklearn import linear_model, datasets, preprocessing
import numpy as np
import matplotlib.pyplot as plt

test_bound = ( total_row *7)//10 #70%的邊界
price_index = total_col -1

### 一般data ###
#除了SalePrice那一欄，其他都要，依據test_bound(70% 30%) 分割資料
x_train = np.array( pd.DataFrame( data, columns = data.columns[0:price_index])[:test_bound] )
x_test  = np.array( pd.DataFrame( data, columns = data.columns[0:price_index])[test_bound:] )
#只取SalePrice那一欄，依據test_bound(70% 30%) 分割資料
y_train =  np.array( data[ data.columns[price_index] ][:test_bound] )
y_test  =  np.array( data[ data.columns[price_index] ][test_bound:] )

### normalize的data ###
#先產生normalize 的data
normalize_data = pd.DataFrame( preprocessing.scale(data) )
#除了SalePrice那一欄，其他都要，依據test_bound(70% 30%) 分割資料
normal_x_train = pd.DataFrame( normalize_data, columns = normalize_data.columns[0:price_index])[:test_bound]
normal_x_test  = pd.DataFrame( normalize_data, columns = normalize_data.columns[0:price_index])[test_bound:]
```

---
#### [ 5. 依據最佳結果的模型,對預測集資料繪製出預測房價 vs 實際房價之散佈(scatter plot)圖 ]
##### 分別使用 Linear、Ridge、及 Lasso 三種 regression 模型預測 Rating,
##### 並使用 MSE (Mean-Squared-Error)作為預測準確度的指標。
##### 比較那一種模型較佳。
---
##### 跑圖，下面這個 cell 是 Linear Regression

```python
#選擇模型 Linear Regression
linear_regr = linear_model.LinearRegression().fit(x_train, y_train)
y_pred = linear_regr.predict(x_test)

#print('Intercept: \n', linear_regr.intercept_)
#print('Coefficients: \n', linear_regr.coef_)

# 計算 mean square error (MSE)
N = len(y_test)
ln_mse_test = np.sum((np.array(y_test).flatten() - np.array(y_pred).flatten())**2) /N 
ln_rmse_test = np.sqrt( ln_mse_test )

print("=====================================================================")
print("Linear MSE TEST:", ln_mse_test)
print("開根號後代表預測房價與實際房價的價格差距約:", int(ln_rmse_test), "元")
print("=====================================================================")

plt.xlabel('Real price')
plt.ylabel('Predict price')
plt.title('Compare value:')
plt.scatter(y_test, y_pred, color='blue', label="Linear")
plt.legend()

plt.show
```

---
##### 跑圖，下面這個 cell 是 Ridge regression
##### Ridge 的 α 選用預設值 1

```python
#選擇模型 Ridge Regression
ridge_regr = linear_model.Ridge(alpha = 1).fit(x_train, y_train)
y_pred = ridge_regr.predict(x_test)

#print('Intercept: \n', ridge_regr.intercept_)
#print('Coefficients: \n', ridge_regr.coef_)

# 計算 mean square error (MSE)
N = len(y_test)
rd_mse_test = np.sum((np.array(y_test).flatten() - np.array(y_pred).flatten())**2) /N 
rd_rmse_test = np.sqrt( rd_mse_test )

print("=====================================================================")
print("Ridge MSE TEST:", rd_mse_test)
print("開根號後代表預測房價與實際房價的價格差距約:", int(rd_rmse_test), "元")
print("=====================================================================")

plt.xlabel('Real price')
plt.ylabel('Predict price')
plt.title('Compare value:')
plt.scatter(y_test, y_pred, color='green', label="Ridge")
plt.legend()

plt.show
```

---
##### 跑圖，下面這個 cell 是 Lasso regression
##### Lasso 的 α 選用預設值 1

```python
#選擇模型 Lasso Regression
lasso_regr = linear_model.Lasso(alpha = 1.0, max_iter=3000).fit(x_train, y_train)
y_pred = lasso_regr.predict(x_test)

#print('Intercept: \n', lasso_regr.intercept_)
#print('Coefficients: \n', lasso_regr.coef_)

# 計算 mean square error (MSE)
N = len(y_test)
ls_mse_test = np.sum((np.array(y_test).flatten() - np.array(y_pred).flatten())**2) /N 
ls_rmse_test = np.sqrt( ls_mse_test )

print("=====================================================================")
print("Lasso MSE TEST: ", ls_mse_test)
print("開根號後代表預測房價與實際房價的價格差距約:", int(ls_rmse_test), "元")
print("=====================================================================")

plt.xlabel('Real price')
plt.ylabel('Predict price')
plt.title('Compare value:')
plt.scatter(y_test, y_pred, color='red', label="Lasso")
plt.legend()

plt.show
```

##### 從上面三個Cell的輸出，可以看到
##### 對於我們的這份資料，三種模型的MSE是很接近的
##### Ridge Regression 有略好一點點

```python
#選擇模型 Ridge Regression
ridge_regr = linear_model.Ridge(alpha = 1)

#先算沒刪極端點的
ridge_regr.fit(x_train, y_train)
ori_y_pred = ridge_regr.predict(x_test)

# 計算 mean square error (MSE)
N = len(y_test)
ori_mse_test = np.sum((np.array(y_test).flatten() - np.array(ori_y_pred).flatten())**2) /N 
ori_rmse_test = np.sqrt( ori_mse_test )

#再算有刪極端點的
ridge_regr.fit(new_x_train, new_y_train)
del_y_pred = ridge_regr.predict(new_x_test)

# 計算 mean square error (MSE)
N = len(new_y_test)
del_mse_test = np.sum((np.array(new_y_test).flatten() - np.array(del_y_pred).flatten())**2) /N 
del_rmse_test = np.sqrt( del_mse_test )

print("=====================================================================")
print("Origin MSE TEST: ", ori_mse_test)
print("原本的：預測房價與實際房價的價格差距約:", int(ori_rmse_test), "元")
print("----------")
print("Delete MSE TEST: ", del_mse_test)
print("刪極端點的：預測房價與實際房價的價格差距約:", int(del_rmse_test), "元")
print("=====================================================================")

plt.title('Origin vs After-delete')
plt.subplot(121)
plt.scatter(y_test, ori_y_pred, color='blue', label="Origin   ")
plt.legend()

plt.subplot(122)
plt.scatter(new_y_test, del_y_pred, color='green', label="Delete")
plt.legend()

plt.show
```
